# Creating item properties for the books database
If the code has already been run once, then the properties already exist in the new database. I have added the first code cell below to delete the propeties before creating them again, so that the code can be run multiple times without errors (via the "run all cells" command).

```python
from recombee_api_client.api_requests import DeleteItemProperty

properties = ['title', 'genre', 'author', 'sales']
for prop in properties:
    try:
        client.send(DeleteItemProperty(prop))
    except Exception as e:
        pass  # if the code hasn't been run before, the properties don't exist yet so their deletion is not possible - thus skipping
```

```python
from recombee_api_client.api_requests import AddItemProperty
requests = [
    AddItemProperty('title', 'string'),
    AddItemProperty('genre', 'string'),
    AddItemProperty('author', 'string'),
    AddItemProperty('sales', 'int')]

for req in requests:
    client.send(req)
```

```python
import pandas as pd
df = pd.read_csv('best-selling-books.csv')

df['id'] = range(1, len(df) + 1)

from recombee_api_client.api_requests import SetItemValues, Batch

requests = []
for _, row in df.iterrows():
    requests.append(SetItemValues(
        str(row['id']),
        {
            'title': str(row['Book']),
            'genre': str(row['Genre']) if pd.notnull(row['Genre']) else '',
            'author': str(row['Author(s)']) if pd.notnull(row['Author(s)']) else '',
            'sales': int(float(row['Approximate sales in millions'])) if pd.notnull(row['Approximate sales in millions']) else 0
        },
        cascade_create=True
    ))

client.send(Batch(requests))
```
